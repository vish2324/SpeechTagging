```python
#read training file 
#filename = "train"
def read_training(filename):
    train = []
    f = open(filename, 'r', encoding="utf-8")
    while True:
        line = f.readline()
        if(line == ''):
            break
        if(line == '\n'):
            train.append(("",""))
            continue
        temp = line.replace('\n','').split(' ')
        train.append(tuple(temp))

    f.close()
    #print(train)
        
    return train
  
read_training("EN/train")
```

```python
#filename = "dev.in"
def read_input(filename):
    inp = []
    f = open(filename, 'r', encoding="utf-8")
    while True:
        line = f.readline()
        if(line == ''):
            break
        if(line == '\n'):
            inp.append('')
            continue
        inp.append(line.replace('\n',''))
        
    f.close()
    #print(inp)
    
    return inp

#read_input("EN/dev.in")
read_input("EN/test.in")
```

Calculating the emission parameters for the training set

```python
#e(x|y)=(count_x_given_y)/(count_y)
# y is the number of tags and x are the words
#count number of times y appears in the seq -- counting the number of tags
#(count_x_given_y) -- is the number of times tag y emits the word x

from collections import Counter

def part2_emission(filename_train):
    train = read_training(filename_train)
    count_x_given_y_data = {}
    emission = {}
    s = []
    k = 1
    #calculating all the count_y
    count_y_data = Counter(sublist[1] for sublist in train if sublist[1] != "")
    #print(count_y_data)

    count = 0

    #calculating all the x given y
    for key in count_y_data:
        y = key
        #print(y)
        val_y = (count_y_data[key] + k)
        #print(val_y)
        for item in train:
            if item[1] == key:
                s.append(item[0])
        count = len(s)
        cnt = Counter()
        for word in s:
            cnt[word] += 1
        data = cnt
        #print(data) 
        for key in data:
            val_x = data[key] 
            x = key 
            key_ = (x,y)
            emission[key_] = val_x/val_y
        
        unk_key_ = ("#UNK#",y)
        emission[unk_key_] = k/val_y
        #print(unk_key_, emission[unk_key_])
        
        s = []
        
    #print(len(emission))
    return emission

part2_emission("EN/train")
```

```python
def part3_transition(filename_train):
    tags = {}  # number of times the y(i-1)-th tag appears in the training set
    trans = {}  # number of times the transition y(i-1) --> y(i) appears in the training set
    transition = {}
    train = read_training(filename_train)
    #print(train)

    p_State = '' # previous state
    c_State = 'START' # current state

    #Count 
    for item in train:
        #print(item)
        if (c_State != 'STOP'):
            p_State = c_State 
        else:
            p_State ='START'  # y(i-1)

        if item != ("",""):  # if still within sentence
            c_State = item[1] 
        else:  # end of sentence
            if p_State == 'START': break
            c_State = 'STOP'  

        if p_State not in tags:  
            tags[p_State] = 1
            trans[p_State] = {c_State: 1}
        else:
            tags[p_State] += 1
            if c_State not in trans[p_State]: trans[p_State][c_State] = 1
            else:trans[p_State][c_State] += 1
        
        #print(item, c_State, p_State)
        #print(trans)

    # Transition Parameters
    #print(trans)
    for tag in trans:
        #print(tag)
        transition[tag] = {}
        #print(transition.keys())
        for i in trans[tag]:
            transition[tag][i] = trans[tag][i]/tags[tag]
            #print(trans[tag][i], tags[tag])
    
    '''
    # check total probability of each state (should be equal to 1, may be slightly off due to rounding)
    for key in transition.keys():
        #print(key, transition[key])
        total_prob = 0
        for tag in transition[key].keys():
            #print(key, tag, transition[key][tag])
            total_prob += transition[key][tag]
        print("prob =", total_prob)
    '''
    
    #print(transition.keys())
    #print(len(transition.keys()))
            
    return transition
  
part3_transition("EN/train")
```

```python
def part3_viterbi(sentence, emission , transition):
  
    # nodes: x(0) [START], x(1), ..., x(n), x(n+1) [STOP]
    # possible states/tags for each node: 18
    
    #print("emission probabilities:", emission) # emission probability table
    #print("transition probabilities:", transition) # transition probability table
    
    # list of all possible tags for each word
    tags = []
    for tag in transition:
        if tag != "START" and tag != "STOP":
            tags.append(tag)
    #print(len(tags), tags)
      
    n = len(sentence) # number of words in sentence
    
    prob = {}
    
    # iterate through each level/word
    
    #print("x(0) = START")
    
    for index in range(0,n):
        word = sentence[index]
        i = index + 1
        #print("x("+str(i)+"):", word)
        
        prob_xi = {}
        
        # base case (i=1)
        if i == 1:
            #print("Base Case: x(1) =", sentence[0])

            for curr_tag in tags:
                if curr_tag in transition["START"] and (word,curr_tag) in emission:
                    prob_xi[curr_tag] = ("START",transition["START"][curr_tag] * emission[(word,curr_tag)])
                    #print("START", curr_tag, "trans:", transition["START"][curr_tag], "emiss:", emission[(word,curr_tag)])

            #print(prob_xi)
            prob["x"+str(i)] = prob_xi
            #print(prob)
        
        # moving forward recursively (i=2,...,n)
        else:
            #print("Moving forward recursively: x(" + str(i) + ") =", sentence[i-1])
        
            for curr_tag in tags:
                possible_parents = {}
              
                found_in_train = False
            
                for key in emission:   
                    if word == key[0]:
                        found_in_train = True

                if found_in_train == False:
                    word = "#UNK#"
                
                for prev_tag in tags:
                    #print(prev_tag, curr_tag, "x"+str(i), word)
                  
                    if curr_tag in transition[prev_tag] and (word,curr_tag) in emission and prev_tag in prob["x"+str(i-1)]:
                        possible_parents[prev_tag] = prob["x"+str(i-1)][prev_tag][1] * transition[prev_tag][curr_tag] * emission[(word,curr_tag)]
                        #prob_xi[curr_tag] = transition[prev_tag][curr_tag] * emission[(word,curr_tag)]
                        #print(prev_tag, curr_tag, "trans:", transition[prev_tag][curr_tag], "emiss:", emission[(word,curr_tag)])
                
                #print(possible_parents)
                
                if possible_parents: # there is at least one possible parent
                    max_parent = max(possible_parents, key=(lambda key: possible_parents[key]))
                    #print("MAX", max_parent, possible_parents[max_parent])
                    prob_xi[curr_tag] = (max_parent, possible_parents[max_parent])
                    
            #print(prob_xi)
            prob["x"+str(i)] = prob_xi
            #print(prob)      
                    
    # final case (i=n+1)
    #print("Final case: x("+str(n+1)+") = STOP")
    prob_xi = {}
    possible_parents = {}
    
    for prev_tag in tags:
        #print(prev_tag, curr_tag, "x"+str(i), word)
                  
        if "STOP" in transition[prev_tag] and prev_tag in prob["x"+str(n)]:
            possible_parents[prev_tag] = prob["x"+str(n)][prev_tag][1] * transition[prev_tag]["STOP"]
            #prob_xi[curr_tag] = transition[prev_tag][curr_tag] * emission[(word,curr_tag)]
            #print(prev_tag, curr_tag, "trans:", transition[prev_tag][curr_tag], "emiss:", emission[(word,curr_tag)])
                
        #print(possible_parents)
                
        if possible_parents: # there is at least one possible parent
            max_parent = max(possible_parents, key=(lambda key: possible_parents[key]))
            #print("MAX", max_parent, possible_parents[max_parent])
            prob_xi["STOP"] = (max_parent, possible_parents[max_parent])
                    
    #print(prob_xi)
    prob["x"+str(n+1)] = prob_xi
    #print(prob)
    
    # backtracking
    seq = ["STOP"]
    curr_node = "STOP"
    for i in range(n+1,0,-1):
        #print("x"+str(i+1))
        #print("x"+str(i), prob["x"+str(i)])
        
        prev_node = curr_node
        
        curr_node = prob["x"+str(i)][curr_node][0]
        #print("Optimal parent of", prev_node, "is", curr_node)
        seq.append(curr_node)
        #print(seq)
        
    seq = seq[::-1]
    #print("Most likely sequence of tags:", seq)
        
    return seq

test_sentence = ["NO", "Saints", "R", ".", "Buch", "might", "come", "back", "n", "play", "vs", "Seahawks", "on", "Sunday", "??"]
part3_viterbi(test_sentence, part2_emission("EN/train"), part3_transition("EN/train"))
```

```python
# address numerical underflow issue: the Viterbi algorithm in log-space
def part3_viterbi_num_stable(sentence, emission , transition):
  
    # nodes: x(0) [START], x(1), ..., x(n), x(n+1) [STOP]
    # possible states/tags for each node: 18
    
    #print("emission probabilities:", emission) # emission probability table
    #print("transition probabilities:", transition) # transition probability table
    
    # list of all possible tags for each word
    tags = []
    for tag in transition:
        if tag != "START" and tag != "STOP":
            tags.append(tag)
    #print(len(tags), tags)
      
    n = len(sentence) # number of words in sentence
    
    prob = {}
    
    # iterate through each level/word
    
    #print("x(0) = START")
    
    for index in range(0,n):
        word = sentence[index]
        i = index + 1
        #print("x("+str(i)+"):", word)
        
        prob_xi = {}
        
        # base case (i=1)
        if i == 1:
            #print("Base Case: x(1) =", sentence[0])

            # take the log of the probabilities (note that log(A*B) = log(A) + log(B))
            # for probabilities that are zero, log(zero) is replaced with some representation of negative infinity (in this case, -sys.maxsize-1)
            for curr_tag in tags:
                if curr_tag not in transition["START"] and (word,curr_tag) not in emission:
                    prob_xi[curr_tag] = ("START",(-sys.maxsize-1) + (-sys.maxsize-1))
                elif curr_tag not in transition["START"]:
                    prob_xi[curr_tag] = ("START",(-sys.maxsize-1) + math.log(emission[(word,curr_tag)]))
                elif (word,curr_tag) not in emission:
                    prob_xi[curr_tag] = ("START",math.log(transition["START"][curr_tag]) + (-sys.maxsize-1))
                else:    
                    prob_xi[curr_tag] = ("START",math.log(transition["START"][curr_tag]) + math.log(emission[(word,curr_tag)]))
                    #print("START", curr_tag, "trans:", transition["START"][curr_tag], "emiss:", emission[(word,curr_tag)])

            #print(prob_xi)
            prob["x"+str(i)] = prob_xi
            #print(prob)
        
        # moving forward recursively (i=2,...,n)
        else:
            #print("Moving forward recursively: x(" + str(i) + ") =", sentence[i-1])
        
            for curr_tag in tags:
                possible_parents = {}
              
                found_in_train = False
            
                for key in emission:   
                    if word == key[0]:
                        found_in_train = True

                if found_in_train == False:
                    word = "#UNK#"
                
                for prev_tag in tags:
                    #print(prev_tag, curr_tag, "x"+str(i), word)
                    
                    if prev_tag in prob["x"+str(i-1)]:
                        if curr_tag not in transition[prev_tag] and (word,curr_tag) not in emission:
                            possible_parents[prev_tag] = prob["x"+str(i-1)][prev_tag][1] + (-sys.maxsize-1) + (-sys.maxsize-1)
                        elif curr_tag not in transition[prev_tag]:
                            possible_parents[prev_tag] = prob["x"+str(i-1)][prev_tag][1] + (-sys.maxsize-1) + math.log(emission[(word,curr_tag)])
                        elif (word,curr_tag) not in emission:
                            possible_parents[prev_tag] = prob["x"+str(i-1)][prev_tag][1] + math.log(transition[prev_tag][curr_tag]) + (-sys.maxsize-1)
                        else:
                            #print("part 3 emission: ",emission[(word,curr_tag)])
                            possible_parents[prev_tag] = prob["x"+str(i-1)][prev_tag][1] + math.log(transition[prev_tag][curr_tag]) + math.log(emission[(word,curr_tag)])
                            #print(prev_tag, curr_tag, "trans:", transition[prev_tag][curr_tag], "emiss:", emission[(word,curr_tag)])
                
                #print(possible_parents)
                
                if possible_parents: # there is at least one possible parent
                    max_parent = max(possible_parents, key=(lambda key: possible_parents[key]))
                    #print("MAX", max_parent, possible_parents[max_parent])
                    prob_xi[curr_tag] = (max_parent, possible_parents[max_parent])
                    
            #print(prob_xi)
            prob["x"+str(i)] = prob_xi
            #print(prob)      
                    
    # final case (i=n+1)
    #print("Final case: x("+str(n+1)+") = STOP")
    prob_xi = {}
    possible_parents = {}
    
    for prev_tag in tags:
        #print(prev_tag, curr_tag, "x"+str(i), word)
        
        if prev_tag in prob["x"+str(n)]:
            if "STOP" not in transition[prev_tag]:
                possible_parents[prev_tag] = prob["x"+str(n)][prev_tag][1] + (-sys.maxsize-1)
            else:
                possible_parents[prev_tag] = prob["x"+str(n)][prev_tag][1] + math.log(transition[prev_tag]["STOP"])
                
        #print(possible_parents)
                
        if possible_parents: # there is at least one possible parent
            max_parent = max(possible_parents, key=(lambda key: possible_parents[key]))
            #print("MAX", max_parent, possible_parents[max_parent])
            prob_xi["STOP"] = (max_parent, possible_parents[max_parent])
                    
    #print(prob_xi)
    prob["x"+str(n+1)] = prob_xi
    #print(prob)
    
    # backtracking
    seq = ["STOP"]
    curr_node = "STOP"
    for i in range(n+1,0,-1):
        #print("x"+str(i+1))
        #print("x"+str(i), curr_node, prob["x"+str(i)])
        
        prev_node = curr_node
        
        curr_node = prob["x"+str(i)][curr_node][0]
        #print("Optimal parent of", prev_node, "is", curr_node)
        seq.append(curr_node)
        #print(seq)
        
    seq = seq[::-1]
    #print("Most likely sequence of tags:", seq)
        
    return seq

test_sentence = ["NO", "Saints", "R", ".", "Buch", "might", "come", "back", "n", "play", "vs", "Seahawks", "on", "Sunday", "??"]
#test_sentence = ['@iians', 'Id', 'love', 'to', 'see', 'IDWT', 'live', 'from', 'CES', '.', 'And', 'one', 'day', 'I', 'hope', 'to', 'go', 'to', 'ces', 'myself', '.']
part3_viterbi_num_stable(test_sentence, part2_emission("EN/train"), part3_transition("EN/train"))
```

```python
# second-order dependencies
def part4_transition(filename_train):
    trans_den = {}  # number of times the transition y(i-2) --> y(i-1) appears in the training set
    trans_num = {}  # number of times the transition y(i-2) --> y(i-1) --> y(i) appears in the training set
    transition = {}
    train = read_training(filename_train)
    #print(train)

    p2_State = '' # previous previous state y(i-2)
    p1_State = 'START' # previous state y(i-1)
    c_State = 'START' # current state y(i)

    #Count 
    for item in train:
        #print(item)
        if (c_State != 'STOP'):
            p2_State = p1_State
            p1_State = c_State
        else:
            p2_State = 'START'  # y(i-2)
            p1_State = 'START'  # y(i-1)

        if item != ("",""):  # if still within sentence
            c_State = item[1]
        else:  # end of sentence
            c_State = 'STOP'

        if p2_State not in trans_den or p2_State not in trans_num:
            #print(trans_den)
            #print(trans_num)
            if p2_State not in trans_den:
                trans_den[p2_State] = {p1_State: 1}
            if p2_State not in trans_num:
                trans_num[p2_State] = {p1_State: {c_State: 1}}
        else:
            # add count to denominator
            if p1_State not in trans_den[p2_State]:
                trans_den[p2_State][p1_State] = 1
            else:
                trans_den[p2_State][p1_State] += 1
                
            # add count to numerator
            if p1_State not in trans_num[p2_State]:
                trans_num[p2_State][p1_State] = {c_State: 1}
            else:
                if c_State not in trans_num[p2_State][p1_State]:
                    trans_num[p2_State][p1_State][c_State] = 1
                else:
                    trans_num[p2_State][p1_State][c_State] += 1
        
        #print(item, p1_State, p2_State, c_State)
        #print(trans_num[p2_State][p1_State][c_State], trans_den[p2_State][p1_State])

    # Transition Parameters
    #print(trans_den)
    #print(trans_num)
    for p2 in trans_num:
        for p1 in trans_num[p2]:
            #print(p1 in trans_den[p2].keys())
            #print(p1, trans_den[p2].keys())
            for c in trans_num[p2][p1]:
                #print("Numerator =", trans_num[p2][p1][c])
                #print("Denominator =", trans_den[p2][p1])
                if p2 not in transition:
                    transition[p2] = {p1: {c: trans_num[p2][p1][c]/trans_den[p2][p1]}}
                else:
                    if p1 not in transition[p2]:
                        transition[p2][p1] = {c: trans_num[p2][p1][c]/trans_den[p2][p1]}
                    else:
                        if c not in transition[p2][p1]:
                            transition[p2][p1][c] = trans_num[p2][p1][c]/trans_den[p2][p1]
                #print(p2, p1, c, transition[p2][p1][c])
    
    #print(transition)
    
    '''
    # check total probability of each state (should be equal to 1, may be slightly off due to rounding)
    for p2 in transition.keys():
        for p1 in transition[p2].keys():
            #print(p2, p1, transition[p2][p1])
            total_prob = 0
            for c in transition[p2][p1].keys():
                #print(p2, p1, c, transition[p2][p1][c])
                total_prob += transition[p2][p1][c]
            print("prob =", total_prob)
    '''
    
    #print(transition.keys())
    #print(len(transition.keys()))
           
    return transition
  
part4_transition("EN/train")
```

```python
# second-order dependencies
# address numerical underflow issue: the Viterbi algorithm in log-space
def part4_viterbi_num_stable(sentence, emission , transition):
  
    # nodes: x(0) [START], x(1), ..., x(n), x(n+1) [STOP]
    # possible states/tags for each node: 18
    
    #print("emission probabilities:", emission) # emission probability table
    #print("transition probabilities:", transition) # transition probability table
    
    # list of all possible tags for each word
    tags = []
    for tag in transition:
        if tag != "START" and tag != "STOP":
            tags.append(tag)
    #print(len(tags), tags)
      
    n = len(sentence) # number of words in sentence
    
    prob = {}
    
    # iterate through each level/word
    
    #print("x(0) = START")
    
    for index in range(0,n):
        word = sentence[index]
        i = index + 1
        #print("x("+str(i)+"):", word)
        
        prob_xi = {}
        
        found_in_train = False
        
        for key in emission:   
            if word == key[0]:
                found_in_train = True

        if found_in_train == False:
            word = "#UNK#"
        
        # base case (i=1)
        if i == 1:
            #print("Base Case: x(1) =", sentence[0])

            # take the log of the probabilities (note that log(A*B) = log(A) + log(B))
            # for probabilities that are zero, log(zero) is replaced with some representation of negative infinity (in this case, -sys.maxsize-1)
            for curr_tag in tags:
                if curr_tag not in transition["START"]["START"] and (word,curr_tag) not in emission:
                    prob_xi[curr_tag] = ("START", (-sys.maxsize-1) + (-sys.maxsize-1))
                elif curr_tag not in transition["START"]["START"]:
                    prob_xi[curr_tag] = ("START", (-sys.maxsize-1) + math.log(emission[(word,curr_tag)]))
                elif (word,curr_tag) not in emission:
                    prob_xi[curr_tag] = ("START", math.log(transition["START"]["START"][curr_tag]) + (-sys.maxsize-1))
                else:    
                    prob_xi[curr_tag] = ("START", math.log(transition["START"]["START"][curr_tag]) + math.log(emission[(word,curr_tag)]))

            #print(prob_xi)
            prob["x"+str(i)] = prob_xi
            #print(prob)
        
        # base case (i=2)
        elif i == 2:
            #print("Base Case: x(2) =", sentence[1])

            # take the log of the probabilities (note that log(A*B) = log(A) + log(B))
            # for probabilities that are zero, log(zero) is replaced with some representation of negative infinity (in this case, -sys.maxsize-1)
            for curr_tag in tags: # i
                possible_parents = {}
                
                for prev1_tag in tags: # i-1
                    
                    #if curr_tag not in transition["START"][prev1_tag] and (word,curr_tag) not in emission:
                    if prev1_tag not in transition["START"] and (word,curr_tag) not in emission:
                        #prob_xi[curr_tag] = (prev1_tag, prob["x"+str(i-1)][prev1_tag][1] + (-sys.maxsize-1) + (-sys.maxsize-1))
                        possible_parents[prev1_tag] = prob["x"+str(i-1)][prev1_tag][1] + (-sys.maxsize-1) + (-sys.maxsize-1)
                    elif prev1_tag not in transition["START"]:
                        #prob_xi[curr_tag] = (prev1_tag, prob["x"+str(i-1)][prev1_tag][1]+ (-sys.maxsize-1) + math.log(emission[(word,curr_tag)]))
                        possible_parents[prev1_tag] = prob["x"+str(i-1)][prev1_tag][1]+ (-sys.maxsize-1) + math.log(emission[(word,curr_tag)])
                    elif (word,curr_tag) not in emission:
                        if curr_tag not in transition["START"][prev1_tag]:
                            #prob_xi[curr_tag] = (prev1_tag, prob["x"+str(i-1)][prev1_tag][1] + (-sys.maxsize-1) + (-sys.maxsize-1))
                            possible_parents[prev1_tag] = prob["x"+str(i-1)][prev1_tag][1] + (-sys.maxsize-1) + (-sys.maxsize-1)
                        else:
                            #prob_xi[curr_tag] = (prev1_tag, prob["x"+str(i-1)][prev1_tag][1]+ math.log(transition["START"][prev1_tag][curr_tag]) + (-sys.maxsize-1))
                            possible_parents[prev1_tag] = prob["x"+str(i-1)][prev1_tag][1]+ math.log(transition["START"][prev1_tag][curr_tag]) + (-sys.maxsize-1)
                    else:
                        if curr_tag not in transition["START"][prev1_tag]:
                            #prob_xi[curr_tag] = (prev1_tag, prob["x"+str(i-1)][prev1_tag][1] + (-sys.maxsize-1) + math.log(emission[(word,curr_tag)]))
                            possible_parents[prev1_tag] = prob["x"+str(i-1)][prev1_tag][1] + (-sys.maxsize-1) + math.log(emission[(word,curr_tag)])
                        else:
                            #prob_xi[curr_tag] = (prev1_tag, prob["x"+str(i-1)][prev1_tag][1] + math.log(transition["START"][prev1_tag][curr_tag]) + math.log(emission[(word,curr_tag)]))
                            possible_parents[prev1_tag] = prob["x"+str(i-1)][prev1_tag][1] + math.log(transition["START"][prev1_tag][curr_tag]) + math.log(emission[(word,curr_tag)])

                #print(possible_parents)
                
                if possible_parents: # there is at least one possible parent
                    max_parent = max(possible_parents, key=(lambda key: possible_parents[key]))
                    #print("MAX", max_parent, possible_parents[max_parent])
                    prob_xi[curr_tag] = (max_parent, possible_parents[max_parent])
                    
            #print(prob_xi)
            prob["x"+str(i)] = prob_xi
            #print(prob)
        
        # moving forward recursively (i=2,...,n)
        else:
            #print("Moving forward recursively: x(" + str(i) + ") =", sentence[i-1])
        
            for curr_tag in tags: # i
                possible_parents = {}
                
                for prev1_tag in tags: # i-1
                    for prev2_tag in tags: # i-2
                        #print(prev2_tag, prev1_tag, curr_tag, "x"+str(i), word)
                        if prev1_tag in prob["x"+str(i-1)]:
                            # if curr_tag not in transition[prev2_tag][prev1_tag] and (word,curr_tag) not in emission:
                            if prev1_tag not in transition[prev2_tag] and (word,curr_tag) not in emission:
                                possible_parents[prev1_tag] = prob["x"+str(i-1)][prev1_tag][1] + (-sys.maxsize-1) + (-sys.maxsize-1)
                            elif prev1_tag not in transition[prev2_tag]:
                                possible_parents[prev1_tag] = prob["x"+str(i-1)][prev1_tag][1] + (-sys.maxsize-1) + math.log(emission[(word,curr_tag)])
                            elif (word,curr_tag) not in emission:
                                if curr_tag not in transition[prev2_tag][prev1_tag]:
                                    possible_parents[prev1_tag] = prob["x"+str(i-1)][prev1_tag][1] + (-sys.maxsize-1) + (-sys.maxsize-1)
                                else:
                                    possible_parents[prev1_tag] = prob["x"+str(i-1)][prev1_tag][1] + math.log(transition[prev2_tag][prev1_tag][curr_tag]) + (-sys.maxsize-1)
                            else:
                                if curr_tag not in transition[prev2_tag][prev1_tag]:
                                    possible_parents[prev1_tag] = prob["x"+str(i-1)][prev1_tag][1] + (-sys.maxsize-1) + math.log(emission[(word,curr_tag)])
                                else:
                                    possible_parents[prev1_tag] = prob["x"+str(i-1)][prev1_tag][1] + math.log(transition[prev2_tag][prev1_tag][curr_tag]) + math.log(emission[(word,curr_tag)])
                
                #print(possible_parents)
                
                if possible_parents: # there is at least one possible parent
                    max_parent = max(possible_parents, key=(lambda key: possible_parents[key]))
                    #print("MAX", max_parent, possible_parents[max_parent])
                    prob_xi[curr_tag] = (max_parent, possible_parents[max_parent])
                    
            #print(prob_xi)
            prob["x"+str(i)] = prob_xi
            #print(prob)      
                    
    # final case (i=n+1)
    #print("Final case: x("+str(n+1)+") = STOP")
    prob_xi = {}
    possible_parents = {}
    
    for prev1_tag in tags: # i-1
        for prev2_tag in tags: # i-2
            #print(prev2_tag, prev1_tag, "STOP", "x"+str(i), word)
            if prev1_tag in prob["x"+str(n)]:
                #if "STOP" not in transition[prev2_tag][prev1_tag]:
                if prev1_tag not in transition[prev2_tag]:
                    possible_parents[prev1_tag] = prob["x"+str(n)][prev1_tag][1] + (-sys.maxsize-1)
                else:
                    if "STOP" not in transition[prev2_tag][prev1_tag]:
                        possible_parents[prev1_tag] = prob["x"+str(n)][prev1_tag][1] + (-sys.maxsize-1)
                    else:
                        possible_parents[prev1_tag] = prob["x"+str(n)][prev1_tag][1] + math.log(transition[prev2_tag][prev1_tag]["STOP"])

            #print(possible_parents)

            if possible_parents: # there is at least one possible parent
                max_parent = max(possible_parents, key=(lambda key: possible_parents[key]))
                #print("MAX", max_parent, possible_parents[max_parent])
                prob_xi["STOP"] = (max_parent, possible_parents[max_parent])
                    
    #print(prob_xi)
    prob["x"+str(n+1)] = prob_xi
    #print(prob)
    
    # backtracking
    seq = ["STOP"]
    curr_node = "STOP"
    for i in range(n+1,0,-1):
        #print("x"+str(i+1))
        #print("x"+str(i), curr_node, prob["x"+str(i)])
        
        prev_node = curr_node

        curr_node = prob["x"+str(i)][curr_node][0]

        #print("Optimal parent of", prev_node, "is", curr_node)
        seq.append(curr_node)
        #print(seq)
        
    seq = seq[::-1]
    #print("Most likely sequence of tags:", seq)
    
    return seq

test_sentence = ["NO", "Saints", "R", ".", "Buch", "might", "come", "back", "n", "play", "vs", "Seahawks", "on", "Sunday", "??"]
#test_sentence = ['@iians', 'Id', 'love', 'to', 'see', 'IDWT', 'live', 'from', 'CES', '.', 'And', 'one', 'day', 'I', 'hope', 'to', 'go', 'to', 'ces', 'myself', '.']
part4_viterbi_num_stable(test_sentence, part2_emission("EN/train"), part4_transition("EN/train"))
```

```python
def read_out(filename):
    out = []
    f = open(filename, 'r', encoding="utf-8")
    while True:
        line = f.readline()
        if(line == ''):
            break
        if(line == '\n'):
            out.append(("",""))
            continue
        temp = line.replace('\n','').split(' ')
        out.append(tuple(temp))

    f.close()
    #print(train)
    #train = [(x.lower(),y.lower()) for x,y in train]
    return out
  
read_out("EN/dev.out")
```

```python
# read the output.p3
def read_p3out(filename):
    p3_out = []
    f = open(filename, 'r', encoding="utf-8")
    while True:
        line = f.readline()
        if(line == ''):
            break
        if(line == '\n'):
            p3_out.append(("",""))
            continue
        temp = line.replace('\n','').split(' ')
        p3_out.append(tuple(temp))

    f.close()
    return p3_out
read_p3out("EN/dev.p3.out")
```

```python
def tags_words(filename , filename_out):
    out = read_out(filename)
    p3_out = read_p3out(filename_out)
    tags = []
    words = []
    tags_pred = []
    for item in out:
        words.append(item[0])
        tags.append(item[1])
    
    
    for item in p3_out:
        tags_pred.append(item[1])
        
    
    return words , tags , tags_pred

tags_words("EN/dev.out" , "EN/dev.p3.out")
```

```python
def perceptron(filename , filename_out, emission , transition):
    words , tags , tags_pred = tags_words(filename , filename_out)
    #print(tags)
    for i in range(len(tags)):
        if tags[i] == '':
            #print("blank")
            continue 
        else:
            pair = (words[i] , tags[i])
            pair_pred = (words[i] , tags_pred[i])
            #print("this is actual pair for " , i , ": " , pair)
            #print("this is predicted pair for " , i , ": " , pair_pred)
            if tags[i] != tags_pred[i]:
                
            #chage the emission values
                #checking if there is a emission value for actual pair
                if pair not in emission:
                    #print(pair , " not in emission")
                    emission[pair] = 0
                if pair_pred in emission:
                    #print("actual emission value for " , i , ": " ,emission[pair])
                    #print("predidctrd emission value for " , i , ": " ,emission[pair_pred])
                    emission[pair] = emission[pair] + 1
                    #emission[pair_pred] = emission[pair_pred] -
                    #print("modified actual emission value for " , i , ": " , emission[pair])
                    
                else:
                    #print(pair_pred ," not in emission")
                    #print("actual emission value for " , i , ": " ,emission[pair])
                    emission[pair] = emission[pair] + 1
                    #print("modified actual emission value for " , i , ": " , emission[pair] )
                    
            #change the transition values
                
                if i == 0 or tags[i-1] == '':
                    #print("its the start")
                    transition['START'][tags[i]] = 1 + transition['START'][tags[i]] 
                    #transition['START'][tags_pred[i]] = transition['START'][tags_pred[i]] - 1
                    #print("transistion of actual for " , i ,  ": " , transition['START'][tags[i]])
                    #print("transistion of pred for " , i ,  ": " , transition['START'][tags_pred[i]])
                else:
                    if tags[i] not in transition[tags[i-1]]:
                        transition[tags[i-1]][tags[i]] = 0

                    transition[tags[i-1]][tags[i]] = 1 + transition[tags[i-1]][tags[i]]
                    #transition[tags_pred[i-1]][tags_pred[i]] = value_tran_pred - transition[tags_pred[i-1]][tags_pred[i]]
                    #print("transistion of actual for " , i ,  ": " , transition[tags[i-1]][tags[i]])
                    #print("transistion of pred for " , i ,  ": " , transition[tags_pred[i-1]][tags_pred[i]])
        #print("---------------------------------------")
                    
    return emission , transition


        
perceptron("EN/dev.out" , "EN/dev.p3.out" , part2_emission("EN/train"), part3_transition("EN/train"))  
```
